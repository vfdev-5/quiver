```python
# Python 
import os

# Keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Convolution2D, MaxPooling2D
from keras.optimizers import SGD

# Quiver
import sys
sys.path.append('../')
from quiver_engine import server

tmp_path='examples/tmp'
```

Get CIFAR10 dataset

```python
from keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

labels = [ 'airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck' ] 

plt.figure(figsize=[12,10])
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.xlabel(labels[int(y_train[i])])
    plt.imshow(X_train[i])
```

```python
# CNN : Conv3x3 -> Pool -> Dense
model = Sequential()
model.add(Convolution2D(12, 3, 3, activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))

sgd = SGD(lr=0.01)
model.compile(optimizer=sgd,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
server.launch(model, classes=labels, temp_folder=tmp_path, input_data=X_train[:10])
```
